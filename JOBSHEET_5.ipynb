```python
# Kelas Induk
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")

# Kelas Anak 1
class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    # Method Override
    def terbang(self):
        print(f"{self.nama} terbang tinggi melayang di angkasa.")

    # Method Override
    def bersuara(self):
        print(f"{self.nama} berteriak nyaring!")

#  Kelas Anak 2
class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    # Method Override
    def terbang(self):
        print(f"{self.nama} terbang cepat di antara pepohonan.")

    # Method Override
    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")

# Fungsi Polimorfisme
def demonstrasi_aksi_burung(daftar_burung):
    print("\n=== Demonstrasi Aksi Burung ===")
    for burung in daftar_burung:
        print(f"\n-- Aksi untuk {burung.nama} --")
        # Polimorfisme: Memanggil metode yang sama, hasil berbeda tiap objek
        burung.terbang()
        burung.bersuara()
    print("-" * 30)

# Kode Utama (Main)
if __name__ == "__main__":
    # Inisialisasi Objek
    elang_jawa = Elang("Elang Jawa", 1.5)
    pipit_gereja = Pipit("Pipit Gereja", "Coklat")
    burung_aneh = Burung("Burung Misterius")

    # Masukkan ke dalam list untuk demonstrasi
    koleksi_burung = [elang_jawa, pipit_gereja, burung_aneh]

    # Eksekusi fungsi
    demonstrasi_aksi_burung(koleksi_burung)
```

```python
# Definisi Kelas

class Kucing:
    def bersuara(self):
        print("Kucing: Meow!")

class Anjing:
    def bersuara(self):
        print("Anjing: Guk guk!")

class Bebek:
    def bersuara(self):
        print("Bebek: Kwek kwek!")

class Mobil:
    # Kelas ini TIDAK punya metode bersuara(),
    # untuk mendemonstrasikan error handling pada Duck Typing.
    def jalankan(self):
        print("Mobil: Brummm!")


# Fungsi Polimorfik (Duck Typing)

def buat_suara(objek_yang_bisa_bersuara):
    """
    Fungsi ini tidak peduli tipe objeknya,
    selama objek tersebut memiliki metode .bersuara()
    """
    try:
        # Memanggil metode 'bersuara' jika ada
        objek_yang_bisa_bersuara.bersuara()
    except AttributeError:
        # Menangani jika objek tidak punya metode 'bersuara'
        nama_kelas = type(objek_yang_bisa_bersuara).__name__
        print(f"Peringatan: Objek {nama_kelas} tidak bisa bersuara.")


# Kode Utama

if __name__ == "__main__":
    # Inisialisasi berbagai macam objek
    kucing1 = Kucing()
    anjing1 = Anjing()
    bebek1 = Bebek()
    mobil1 = Mobil()

    daftar_objek = [kucing1, anjing1, bebek1, mobil1]

    print("=== Demonstrasi Duck Typing ===\n")
    for item in daftar_objek:
        buat_suara(item)
```

```python
# Definisi Kelas

class KalkulatorSederhana:
    def __init__(self, nama="Kalkulator"):
        self.nama = nama
        print(f"{self.nama} siap digunakan.")

    def tambah(self, *args):
        """
        Metode ini menjumlahkan semua argumen angka yang diberikan.
        Menerima sejumlah argumen posisi menggunakan *args.
        """
        print(f"\nMemanggil metode tambah dengan argumen: {args}")

        if not args:
            print("Tidak ada angka untuk dijumlahkan.")
            return 0

        total = 0
        valid_input = True

        for angka in args:
            # Melakukan pengecekan tipe data (int atau float)
            if isinstance(angka, (int, float)):
                total += angka
            else:
                print(f"Peringatan: Argumen '{angka}' bukan angka dan akan diabaikan.")
                valid_input = False

        if valid_input:
            print(f"Hasil penjumlahan: {total}")
        else:
            print(f"Hasil penjumlahan (dengan beberapa input diabaikan): {total}")

        return total


# Kode Utama

if __name__ == "__main__":
    # Inisialisasi objek kalkulator
    calc = KalkulatorSederhana("Calc-01")

    print("\n--- Percobaan Penjumlahan ---")

    # 1. Menjumlahkan dua angka
    calc.tambah(5, 10)
    # 2. Menjumlahkan empat angka
    calc.tambah(2, 3, 5, 10)
    # 3. Hanya satu angka
    calc.tambah(100)
    # 4. Tanpa argumen (kosong)
    calc.tambah()
    # 5. Campuran angka dan string (Error handling)
    calc.tambah(1, 2, "tiga", 4, 5.5)
```

```python
# Definisi Kelas

class DataProcessor:
    def __init__(self, processor_id):
        self.processor_id = processor_id
        print(f"Data Processor {self.processor_id} siap.")

    def process(self, *args, **kwargs):
        """
        Metode ini memproses data dengan argumen posisi (*args)
        dan kata kunci (**kwargs) yang jumlahnya variabel.
        """
        print(f"\n--- {self.processor_id} Memproses Data ---")

        # 1. Memproses argumen posisi (*args -> diperlakukan sebagai tuple)
        if args:
            print("Argumen Posisi Diterima:")
            for i, arg in enumerate(args):
                print(f"  args[{i}]: {arg} (tipe: {type(arg).__name__})")
        else:
            print("Tidak ada argumen posisi.")

        # 2. Memproses argumen kata kunci (**kwargs -> diperlakukan sebagai dictionary)
        if kwargs:
            print("Argumen Kata Kunci Diterima:")
            for kunci, nilai in kwargs.items():
                print(f"  {kunci}: {nilai} (tipe: {type(nilai).__name__})")
        else:
            print("Tidak ada argumen kata kunci.")

        print("---------------------------------")


# Kode Utama

if __name__ == "__main__":
    # Membuat instance dari kelas
    processor1 = DataProcessor("DP-001")

    # Panggilan 1: Tanpa argumen tambahan
    print("\nPanggilan 1: Tanpa argumen tambahan")
    processor1.process()
    # Panggilan 2: Hanya argumen posisi
    print("\nPanggilan 2: Hanya argumen posisi")
    processor1.process(100, "Status OK", 99.9, False)
    # Panggilan 3: Hanya argumen kata kunci
    print("\nPanggilan 3: Hanya argumen kata kunci")
    processor1.process(user="admin", level=5, mode="verbose")
    # Panggilan 4: Kombinasi argumen posisi dan kata kunci
    print("\nPanggilan 4: Kombinasi argumen posisi dan kata kunci")
    processor1.process("Task-A", "Task-B", status="Running", priority="High", thread_id=54321)
    # Panggilan 5: Argumen posisi kompleks (List dan Dictionary)
    print("\nPanggilan 5: Argumen posisi kompleks")

    list_ids = ["A1", "B2", "C3"]
    dict_params = {"timeout": 30, "retry": 3}
    processor1.process(list_ids, dict_params, owner="system", enabled=True)
```

```python
import math

# Definisi Kelas dari Hierarki Inheritance
class Bentuk:
    # Metode ini bisa dianggap sebagai 'template' atau interface dasar
    def nama_bentuk(self):
        return "Bentuk Generik"

    # Metode ini wajib di-override oleh subclass
    def hitung_luas(self):
        raise NotImplementedError("Subclass harus mengimplementasikan metode ini")


class Lingkaran(Bentuk):
    def __init__(self, radius):
        self.radius = radius
    # Override metode dari kelas induk
    def nama_bentuk(self):
        return "Lingkaran"
    # Override metode dari kelas induk
    def hitung_luas(self):
        return math.pi * (self.radius ** 2)


class Persegi(Bentuk):
    def __init__(self, sisi):
        self.sisi = sisi

    # Override metode dari kelas induk
    def nama_bentuk(self):
        return "Persegi"
    # Override metode dari kelas induk
    def hitung_luas(self):
        return self.sisi * self.sisi


# Definisi Kelas Lain (Tidak Terkait Inheritance)
class TaplakMeja:
    def __init__(self, panjang, lebar):
        self.panjang = panjang
        self.lebar = lebar

    # Metode dengan nama sama -> 'hitung_luas'
    # Ini memungkinkan TaplakMeja bekerja dengan fungsi polimorfik (Duck Typing)
    def hitung_luas(self):
        return self.panjang * self.lebar

    def deskripsi(self):
        return f"Taplak Meja {self.panjang}x{self.lebar}"


# Fungsi Polimorfik
def tampilkan_info_luas(objek_dengan_luas):
    print("-" * 30)
    try:
        # Mencoba memanggil hitung_luas() - Inti dari Polimorfisme
        luas = objek_dengan_luas.hitung_luas()
        # Mencoba mendapatkan nama (Duck Typing tambahan)
        try:
            nama = objek_dengan_luas.nama_bentuk()
        except AttributeError:
            # Jika tidak ada nama_bentuk(), ambil nama kelasnya saja
            nama = type(objek_dengan_luas).__name__

        print(f"Objek : {nama}")
        print(f"Luas  : {luas:.2f}")

    except AttributeError:
        # Jika objek tidak punya metode .hitung_luas()
        print(f"Peringatan: {type(objek_dengan_luas).__name__} tidak bisa dihitung luasnya.")
    except NotImplementedError as e:
        # Jika metode ada (di kelas induk) tapi belum di-override
        print(f"Error: {e}")


# Kode Utama
if __name__ == "__main__":
    # Inisialisasi berbagai objek
    lingkaran1 = Lingkaran(7)
    persegi1 = Persegi(5)
    taplak1 = TaplakMeja(1.5, 0.8)
    bentuk_dasar = Bentuk()       # Akan memicu NotImplementedError
    string_biasa = "Ini string"    # Akan memicu AttributeError

    daftar_item = [lingkaran1, persegi1, taplak1, bentuk_dasar, string_biasa]

    print("=== Menampilkan Info Luas (Polimorfisme Campuran) ===")
    for item in daftar_item:
        tampilkan_info_luas(item)
```

```python
# Definisi Kelas Burung, Elang, Pipit
class Burung:
    def __init__(self, nama):
        self.nama = nama
    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")
    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")


class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    # Override: Mengganti perilaku terbang induk
    def terbang(self):
        print(f"{self.nama} terbang tinggi melayang di angkasa.")
    # Override: Mengganti perilaku suara induk
    def bersuara(self):
        print(f"{self.nama} berteriak nyaring!")
    # Metode khusus Elang (hanya ada di kelas Elang)
    def berburu(self):
        print(f"{self.nama} sedang mencari mangsa dari ketinggian.")


class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu
    # Override
    def terbang(self):
        print(f"{self.nama} terbang cepat di antara pepohonan.")
    # Override
    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit cit!")
    # Metode khusus Pipit (hanya ada di kelas Pipit)
    def membangun_sarang(self):
        print(f"{self.nama} sedang mengumpulkan ranting untuk sarang.")


# Fungsi Polimorfisme DAN Type Checking (isinstance)
def interaksi_dengan_burung(burung):
    # Mengambil nama kelas dan atribut nama secara aman
    nama_kelas = type(burung).__name__
    nama_objek = getattr(burung, 'nama', 'Objek tidak dikenal')

    print(f"\n--- Berinteraksi dengan {nama_kelas}: {nama_objek} ---")

    # 1. Cek apakah objek adalah instance dari Burung atau turunannya
    if isinstance(burung, Burung):
        # Perilaku polimorfik dasar
        burung.terbang()
        burung.bersuara()
        # 2. Downcasting (menangani perilaku spesifik berdasarkan tipe turunan)
        if isinstance(burung, Elang):
            print("-> Identifikasi: Ini adalah Elang!")
            burung.berburu()
        elif isinstance(burung, Pipit):
            print("-> Identifikasi: Ini adalah Pipit!")
            burung.membangun_sarang()
        else:
            print("-> Identifikasi: Ini adalah burung jenis umum.")
    else:
        # Jika objek BUKAN instance Burung sama sekali
        print("-> Peringatan: Objek ini bukan termasuk jenis Burung.")

    print("-" * 35)


# Kode Utama
if __name__ == "__main__":
    # Membuat berbagai macam objek
    elang_sumatra = Elang("Elang Sumatra", 1.8)
    pipit_rumah = Pipit("Pipit Rumah", "Abu-abu")
    merak = Burung("Merak")
    kucing_tetangga = "Meong"  # Objek string (non-burung)

    koleksi_makhluk = [elang_sumatra, pipit_rumah, merak, kucing_tetangga]

    for makhluk in koleksi_makhluk:
        interaksi_dengan_burung(makhluk)
```

```python
# Kelas Induk
class Komputer:
    def __init__(self, merk, processor, ram_gb):
        self.merk = merk
        self.processor = processor
        self.ram_gb = ram_gb

    def info_spesifikasi(self):
        print(f"Merk           : {self.merk}")
        print(f"Processor      : {self.processor}")
        print(f"RAM            : {self.ram_gb} GB")
    def jalankan_aplikasi(self, nama_aplikasi):
        print(f"[{self.merk}] menjalankan aplikasi: {nama_aplikasi}...")

    # Simulasi Overloading menggunakan *args (Opsi B)
    def install_software(self, *args):
        if not args:
            print(f"[{self.merk}] Tidak ada software yang dipilih untuk diinstal.")
            return

        print(f"\n--- Proses Instalasi di {self.merk} ---")
        for software in args:
            print(f"Menginstall {software} di {self.merk}...")


# Kelas Anak 1 (Laptop)
class Laptop(Komputer):
    def __init__(self, merk, processor, ram_gb, ukuran_layar_inch, berat_kg):
        # Menginisialisasi atribut induk
        super().__init__(merk, processor, ram_gb)
        self.ukuran_layar_inch = ukuran_layar_inch
        self.berat_kg = berat_kg

    # Override metode info_spesifikasi
    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Ukuran Layar   : {self.ukuran_layar_inch} inch")
        print(f"Berat          : {self.berat_kg} Kg")


# Kelas Anak 2 (Desktop)
class Desktop(Komputer):
    def __init__(self, merk, processor, ram_gb, jenis_casing, monitor_external):
        # Menginisialisasi atribut induk
        super().__init__(merk, processor, ram_gb)
        self.jenis_casing = jenis_casing
        self.monitor_external = monitor_external

    # Override metode info_spesifikasi
    def info_spesifikasi(self):
        super().info_spesifikasi()
        status_monitor = "Ya" if self.monitor_external else "Tidak"
        print(f"Jenis Casing   : {self.jenis_casing}")
        print(f"Monitor External: {status_monitor}")


# Fungsi Polimorfisme
def cetak_semua_spesifikasi(daftar_komputer):
    print("\n" + "="*40)
    print("DAFTAR SPESIFIKASI KOMPUTER")
    print("="*40)
    for komputer in daftar_komputer:
        komputer.info_spesifikasi()
        print("-" * 20)


# --- Kode Utama ---
if __name__ == "__main__":
    # 1. Membuat objek Laptop dan Desktop
    laptop_saya = Laptop("Asus ROG", "Intel i9", 32, 15.6, 2.1)
    pc_kantor = Desktop("Dell OptiPlex", "Intel i5", 16, "Tower", True)

    # 2. Masukkan ke dalam list
    koleksi_komputer = [laptop_saya, pc_kantor]
    # 3. Demonstrasi Polimorfisme
    cetak_semua_spesifikasi(koleksi_komputer)
    # 4. Demonstrasi Simulasi Overloading (*args)
    print("\n[Demonstrasi Simulasi Overloading]")

    # Panggilan dengan satu argumen
    laptop_saya.install_software("VS Code")
    # Panggilan dengan banyak argumen
    pc_kantor.install_software("Excel", "Photoshop", "Chrome", "Spotify")
```
