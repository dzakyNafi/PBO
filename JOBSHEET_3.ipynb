```python
class BankAccount:
  def __init__(self, owner, balance):
    self.__owner = owner
    self.__balance = balance
  def deposit(self, amount):
    """Method untuk menambahkan saldo."""
    if amount > 0:
      self.__balance += amount
      print(f"{amount} telah ditambahkan ke akun {self.__owner}.")
    else:
      print("Jumlah deposit harus lebih dari 0.")
  def withdraw(self, amount):
    """Method untuk menarik saldo."""
    if amount <= self.__balance:
      self.__balance -= amount
      print(f"{amount} telah ditarik dari akun {self.__owner}.")
    else:
      print("Saldo tidak mencukupi.")
  def get_balance(self):
    """Method untuk mendapatkan informasi saldo terkini."""
    return self.__balance


# Contoh penggunaan
if __name__ == "__main__":
  # Membuat objek BankAccount dengan owner="Alice" dan balance awal 1000
  alice_account = BankAccount(owner="Alice", balance=1000)

  # Deposit uang
  alice_account.deposit(500) # Berhasil
  alice_account.deposit(-100) # Gagal (validasi)

  # Withdraw uang
  alice_account.withdraw(300) # Berhasil
  alice_account.withdraw(2000) # Gagal (saldo tidak cukup)

  # Mendapatkan saldo
  current_balance = alice_account.get_balance()
  print(f"Saldo terakhir di akun {alice_account._BankAccount__owner}:{current_balance}")

  # Mencoba mengakses atribut 'private' langsung (tidak direkomendasikan) karen akan menyebabkan error disebabkan Python “mangling” nama atribut __balance menjadi_BankAccount__balance
  alice_account.__balance
```

```python
class Employee:
    def __init__(self, name, salary):
        # Atribut privat dengan double underscore
        self.__name = name
        self.__salary = salary

    def update_salary(self, increase):
        """Method untuk menaikkan gaji dengan validasi."""
        if increase > 0:
            self.__salary += increase
            print(f"Gaji telah dinaikkan sebesar {increase}.")
        else:
            print("Nilai kenaikan harus lebih dari 0.")

    def set_salary(self, new_salary):
        """Method untuk mengubah gaji dengan validasi."""
        if new_salary >= 0:
            self.__salary = new_salary
            print(f"Gaji diatur ulang menjadi {new_salary}.")
        else:
            print("Gaji tidak dapat bernilai negatif.")

    def get_salary(self):
        """Method untuk mendapatkan informasi gaji."""
        return self.__salary

    def get_employee_info(self):
        """Menampilkan informasi karyawan."""
        return f"Employee: {self.__name}, Gaji: {self.__salary}"


# ===== Contoh penggunaan =====
if __name__ == "__main__":
    # Membuat objek Employee
    employee1 = Employee("John Doe", 50000)

    # Tampilkan info awal
    print(employee1.get_employee_info())

    # Update gaji
    employee1.update_salary(5000)
    print(f"Gaji setelah kenaikan: {employee1.get_salary()}")

    # Set ulang gaji
    employee1.set_salary(60000)
    print(f"Informasi terbaru: {employee1.get_employee_info()}")

    # Akses langsung (tidak direkomendasikan)
    # print(employee1.__salary)  # akan error
```

```python
class SimpleExample:
    def __init__(self, name):
        """
        Konstruktor: Dipanggil saat objek dibuat.
        Menyimpan nilai 'name' dan mencetak pesan.
        """
        self.name = name
        print(f"Konstruktor: Objek '{self.name}' telah dibuat.")

    def __del__(self):
        """
        Destruktor: Dipanggil saat objek dihapus.
        """
        print(f"Destruktor: Objek '{self.name}' sedang dihapus.")


def main():
    print("Program dimulai.\n")

    # Membuat objek
    obj = SimpleExample("Demo")
    print("Program sedang berjalan...\n")

    # Menghapus objek secara eksplisit
    del obj
    print("Objek telah dihapus secara eksplisit.\n")

    print("Program selesai.")


if __name__ == "__main__":
    main()
```

```python
class FileLogger:
    def __init__(self, filename):
        """
        Konstruktor: Membuka file log untuk menulis pesan.
        """
        self.filename = filename
        try:
            self.file = open(filename, "a")  # mode append
            print(f"File '{filename}' berhasil dibuka untuk logging.")
        except Exception as e:
            self.file = None
            print(f"Gagal membuka file '{filename}': {e}")

    def write_log(self, message):
        """
        Menulis pesan log ke dalam file.
        """
        if self.file:
            self.file.write(message + "\n")
            self.file.flush()  # langsung tulis ke disk
            print(f"Pesan log: '{message}' telah ditulis.")
        else:
            print("File tidak tersedia untuk menulis.")

    def __del__(self):
        """
        Destruktor: Menutup file log saat objek dihapus.
        """
        if hasattr(self, "file") and self.file and not self.file.closed:
            self.file.close()
            print(f"File '{self.filename}' telah ditutup.")


# ===== Contoh penggunaan =====
if __name__ == "__main__":
    logger = FileLogger("application.log")

    logger.write_log("Aplikasi dimulai.")
    logger.write_log("Melakukan operasi A...")
    logger.write_log("Operasi A selesai.")
    logger.write_log("Aplikasi akan segera selesai.")

    # Hapus objek secara eksplisit
    del logger
```

```python
class Calculator:
    def __init__(self, initial_value=0):
        """
        Konstruktor kelas Calculator.
        """
        self.value = initial_value
        print(f"Kalkulator diinisialisasi dengan nilai: {self.value}")

    def add(self, number):
        """
        Menambahkan nilai ke kalkulator.
        """
        self.value += number
        print(f"Setelah penambahan {number}, nilai sekarang adalah: {self.value}")

    def subtract(self, number):
        """
        Mengurangi nilai dari kalkulator.
        """
        self.value -= number
        print(f"Setelah pengurangan {number}, nilai sekarang adalah: {self.value}")

    def reset(self):
        """
        Reset nilai ke 0.
        """
        self.value = 0
        print("Nilai telah direset ke 0.")

    def show_value(self):
        """
        Menampilkan nilai saat ini.
        """
        print(f"Nilai saat ini adalah: {self.value}")


# ===== Contoh penggunaan =====
def main():
    # Objek pertama
    calc1 = Calculator(initial_value=10)
    calc1.add(5)
    calc1.subtract(3)
    calc1.show_value()

    print()

    # Objek kedua
    calc2 = Calculator()
    calc2.add(20)
    calc2.show_value()

    # Penjelasan:
    # setiap objek punya 'self.value' masing-masing (tidak saling mempengaruhi)


if __name__ == "__main__":
    main()
```

```python
class Person:
    def __init__(self, name, age):
        """
        Konstruktor untuk menginisialisasi objek Person.
        """
        self.__name = name
        self.__age = age

    @property
    def name(self):
        """Getter untuk name."""
        return self.__name

    @name.setter
    def name(self, value):
        """Setter untuk name dengan validasi."""
        if not value:
            print("Nama tidak boleh kosong.")
        else:
            self.__name = value

    @property
    def age(self):
        """Getter untuk age."""
        return self.__age

    @age.setter
    def age(self, value):
        """Setter untuk age dengan validasi."""
        if value < 0:
            print("Umur tidak boleh negatif!")
        else:
            self.__age = value


# ===== Contoh penggunaan =====
def main():
    person = Person("Alice", 30)
    print(f"Nama: {person.name}, Umur: {person.age}")

    # Update data
    person.name = "Bob"
    person.age = 35
    print(f"Nama baru: {person.name}, Umur baru: {person.age}")

    # Coba nilai tidak valid
    person.age = -5  # akan ditolak


if __name__ == "__main__":
    main()
```

```python
class Mahasiswa:
    def __init__(self, nama, nilai):
        self.__nama = nama
        self.__nilai = 0
        self.__grade = ''
        self.nilai = nilai  # pakai setter agar validasi jalan

    # ===== Getter & Setter NAMA =====
    @property
    def nama(self):
        return self.__nama

    @nama.setter
    def nama(self, value):
        self.__nama = value

    # ===== Getter & Setter NILAI =====
    @property
    def nilai(self):
        return self.__nilai

    @nilai.setter
    def nilai(self, value):
        if 0 <= value <= 100:
            self.__nilai = value
            self.__update_grade()  # update otomatis
        else:
            print("Nilai harus antara 0 - 100")

    # ===== Method internal untuk update grade =====
    def __update_grade(self):
        if self.__nilai >= 90:
            self.__grade = 'A'
        elif self.__nilai >= 80:
            self.__grade = 'B'
        elif self.__nilai >= 70:
            self.__grade = 'C'
        elif self.__nilai >= 60:
            self.__grade = 'D'
        else:
            self.__grade = 'E'

    # ===== Method tampilkan info =====
    def tampilkan_info(self):
        print(f"Nama Mahasiswa: {self.__nama}")
        print(f"Nilai: {self.__nilai}")
        print(f"Grade: {self.__grade}")

    # ===== Destructor =====
    def __del__(self):
        print(f"Data mahasiswa {self.__nama} telah dihapus dari sistem.")


# ===== Program utama =====
mhs = Mahasiswa("Siti", 87)
mhs.tampilkan_info()

print("\nNilai diubah...")
mhs.nilai = 93
mhs.tampilkan_info()

del mhs
```
