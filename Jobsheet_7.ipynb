Praktikum 01: try...except Dasar dalam Metode Kelas

```python
class KalkulatorAngka:
    def __init__(self):
        print("Kalkulator Angka siap.")

    def bagi(self, angka1, angka2):
        """
        Membagi angka1 dengan angka2.
        Menangani potensi ValueError (jika input bukan angka)
        dan ZeroDivisionError (jika pembagi nol) secara terpisah.
        """
        print(f"\nMencoba membagi '{angka1}' dengan '{angka2}'...")
        try:
            # Konversi input ke float, bisa memicu ValueError
            num1 = float(angka1)
            num2 = float(angka2)

            # Operasi pembagian, bisa memicu ZeroDivisionError
            hasil = num1 / num2
            print(f" -> Hasil pembagian: {hasil}")
            return hasil

        # Menangkap ValueError secara spesifik
        except ValueError:
            print(" -> ERROR: Input tidak valid! Pastikan kedua input adalah angka.")
            return None

        # Menangkap ZeroDivisionError secara spesifik
        except ZeroDivisionError:
            print(" -> ERROR: Tidak bisa melakukan pembagian dengan nol.")
            return None

        # Menangkap exception lain yang mungkin tidak terduga (opsional)
        except Exception as e:
            print(f" -> ERROR Lainnya: Terjadi kesalahan {e}")
            return None

# Kode Utama
if __name__ == "__main__":
    calc = KalkulatorAngka()
    print("--- Tes Kalkulator ---")

    # Uji coba berbagai skenario input
    hasil1 = calc.bagi(100, 5)          # Operasi sukses
    hasil2 = calc.bagi(20, 0)           # Memicu ZeroDivisionError
    hasil3 = calc.bagi("limapuluh", 10) # Memicu ValueError saat konversi
    hasil4 = calc.bagi(50, "sepuluh")   # Memicu ValueError saat konversi

    print("\n--- Ringkasan Hasil --")
    print(f"Hasil 1 (100/5): {hasil1}")
    print(f"Hasil 2 (20/0): {hasil2}")
    print(f"Hasil 3 ('limapuluh'/10): {hasil3}")
    print(f"Hasil 4 (50/'sepuluh'): {hasil4}")
```

Praktikum 03: Menggunakan else dan finally dalam Metode

```python
import os  # file dummy

class FileProcessor:
    def __init__(self, nama_file):
        self.nama_file = nama_file
        self._file_handle = None  # Untuk menyimpan objek file saat dibuka
        print(f"\n-> FileProcessor dibuat untuk file: '{self.nama_file}'")

    def baca_dan_proses(self):
        """Membuka, membaca, dan 'memproses' file, dengan cleanup di finally."""
        print(f"[{self.nama_file}] Mencoba memproses...")
        try:
            # 1. Mencoba membuka file untuk dibaca
            print(f"  [{self.nama_file}] Membuka file...")
            self._file_handle = open(self.nama_file, 'r', encoding="utf-8")  # Mode 'r' untuk baca

            # 2. Jika buka berhasil, coba baca isinya
            print(f"  [{self.nama_file}] Membaca isi file...")
            isi = self._file_handle.read()

            # 3. Simulasi pemrosesan isi
            print(f"  [{self.nama_file}] Memproses isi...")
            panjang_isi = len(isi)
            print(f" -> File berhasil dibaca.")
            print(f" -> Panjang isi: {panjang_isi} karakter.")

        except FileNotFoundError:
            # 4a. Blok ini jalan jika open() gagal karena file tidak ada
            print(f" -> ERROR: File '{self.nama_file}' tidak ditemukan.")

        except UnicodeDecodeError:
            # 4b. Contoh error lain saat membaca file non-teks
            print(f" -> ERROR: Gagal membaca '{self.nama_file}' sebagai teks (mungkin file biner?).")

        except Exception as e:
            # 4c. Menangkap error tak terduga lainnya saat try
            print(f" -> ERROR Lainnya saat memproses file: {type(e).__name__} - {e}")

        else:
            # 5. Blok ini jalan HANYA JIKA try selesai tanpa exception
            print(f" -> PEMROSESAN BERHASIL: Tidak ada error terjadi pada '{self.nama_file}'.")

        finally:
            # 6. Blok ini SELALU jalan setelah try/except/else selesai
            print(f"  [{self.nama_file}] Blok Finally: Melakukan cleanup...")
            if self._file_handle and not self._file_handle.closed:
                print(f"  Menutup file '{self.nama_file}'...")
                self._file_handle.close()
                print(" -> File berhasil ditutup.")
            else:
                # Ini akan terjadi jika file gagal dibuka (FileNotFoundError)
                # atau jika sudah ditutup karena error lain di try
                print(" -> File tidak perlu ditutup (belum dibuka atau sudah ditutup).")

            print(f"[{self.nama_file}] Selesai mencoba proses.")
            print("-" * 30)

# Kode Utama
if __name__ == "__main__":
    NAMA_FILE_SUKSES = "data_penting.txt"
    NAMA_FILE_GAGAL = "file_yang_tidak_ada.log"

    # Membuat file dummy untuk skenario sukses
    try:
        with open(NAMA_FILE_SUKSES, "w", encoding='utf-8') as f:
            f.write("Ini adalah baris pertama.\n")
            f.write("Baris kedua berisi data penting.\n")
        print(f"File dummy '{NAMA_FILE_SUKSES}' berhasil dibuat.")
    except IOError as e:
        print(f"Gagal membuat file dummy '{NAMA_FILE_SUKSES}': {e}")

    # Skenario 1: File ada dan proses berhasil
    print("\n--- Skenario 1: Proses Sukses ---")
    processor1 = FileProcessor(NAMA_FILE_SUKSES)
    processor1.baca_dan_proses()

    # Skenario 2: File tidak ada
    print("\n--- Skenario 2: File Tidak Ditemukan ---")
    processor2 = FileProcessor(NAMA_FILE_GAGAL)
    processor2.baca_dan_proses()

    # Cleanup file dummy setelah selesai digunakan
    try:
        if os.path.exists(NAMA_FILE_SUKSES):
            os.remove(NAMA_FILE_SUKSES)
            print(f"\nFile dummy '{NAMA_FILE_SUKSES}' dihapus.")
    except OSError as e:
        print(f"\nGagal menghapus file dummy '{NAMA_FILE_SUKSES}': {e}")
```

Praktikum 05: Membangkitkan Exception (raise) dalam Metode

```python
import locale
try:
  locale.setlocale (locale. LC_ALL, 'id_ID.UTF-8')
except locale. Error:
  print("Locale id_ID tidak tersedia.")
def format_rupiah (angka):
  try:
    return locale.currency (angka, grouping-True, symbol='Rp')
  except NameError: # Jika locale gagal di-set
    return f"Rp {angka:,.2f}"

class PengelolaAkun:
    def __init__(self, nomor_akun: str, saldo_awal: float = 0):
        self.nomor_akun = nomor_akun

        # Mengatur lokal ke Indonesia untuk format mata uang Rupiah
        try:
            locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
        except locale.Error:
            # Fallback jika environment sistem tidak mendukung UTF-8 ID secara langsung
            locale.setlocale(locale.LC_ALL, '')

        # Aturan diagram halaman 12: Memeriksa saldo_awal, jika < 0 melempar ValueError
        if saldo_awal < 0:
            raise ValueError("Saldo awal tidak boleh kurang dari nol!")

        self._saldo = float(saldo_awal)  # Atribut protected (_)

    def format_rupiah(self, nominal: float) -> str:
        """Fungsi pembantu lokal untuk mengubah float menjadi format mata uang Rupiah."""
        return locale.currency(nominal, grouping=True, symbol=True)

    def get_saldo(self) -> float:
        return self._saldo

    def tarik_tunai(self, jumlah: float) -> bool:
        """
        Melakukan transaksi penarikan tunai dengan validasi exception.
        - Jika jumlah <= 0, melempar ValueError
        - Jika jumlah > saldo, melempar RuntimeError
        """
        try:
            # Aturan diagram halaman 12: Memeriksa nominal jumlah penarikan
            if jumlah <= 0:
                raise ValueError("Jumlah penarikan harus lebih besar dari nol.")

            # Aturan diagram halaman 12: Memeriksa kecukupan saldo akun
            if jumlah > self._saldo:
                raise RuntimeError("Saldo tidak mencukupi untuk melakukan penarikan ini.")

            # Jika semua kondisi lolos validasi
            self._saldo -= jumlah
            print(f"[{self.nomor_akun}] Sukses menarik {self.format_rupiah(jumlah)}. Sisa saldo: {self.format_rupiah(self._saldo)}")
            return True

        except ValueError as e_val:
            print(f"[{self.nomor_akun}] ValueError ditangkap: {e_val}")
            return False
        except RuntimeError as e_run:
            print(f"[{self.nomor_akun}] RuntimeError ditangkap: {e_run}")
            return False

# Pengujian Praktikum 05
if __name__ == "__main__":
    print("=== RUNNING: PRAKTIKUM 05 (REVISED) ===")

    # Inisialisasi akun dengan saldo awal Rp500.000,00
    akun_mhs = PengelolaAkun("NIM-432242", 500000)
    print(f"Saldo Awal: {akun_mhs.format_rupiah(akun_mhs.get_saldo())}")

    # 1. Menguji aturan ValueError (jumlah <= 0)
    print("\n[Test 1] Menarik nominal minus:")
    akun_mhs.tarik_tunai(-25000)

    # 2. Menguji aturan RuntimeError (jumlah > saldo)
    print("\n[Test 2] Menarik melebihi saldo:")
    akun_mhs.tarik_tunai(750000)

    # 3. Menguji transaksi normal yang sukses
    print("\n[Test 3] Menarik nominal valid:")
    akun_mhs.tarik_tunai(150000)
```

Praktikum 06: Kombinasi Menangkap dan Membangkitkan Exception

```python
# 1. Custom Exception Class berdasarkan rancangan Praktikum 06
class KonfigurasiError(Exception):
    def __init__(self, message: str, penyebab_asli: Exception = None):
        super().__init__(message)
        self.penyebab_asli = penyebab_asli  # Menyimpan objek kesalahan dasar dari sistem [cite: 1224]

    def __str__(self) -> str:
        # Menampilkan pesan error custom digabung dengan rincian error aslinya [cite: 1224]
        if self.penyebab_asli:
            return f"{super().__str__()} [Penyebab Internal: {type(self.penyebab_asli).__name__} - {self.penyebab_asli}]"
        return super().__str__()


# 2. Kelas Komponen Bisnis PengelolaSetting
class PengelolaSetting:
    def __init__(self, file_config: str):
        self.file_config = file_config
        self._config = {}  # Atribut protected dictionary internal
        self._load_config()

    def _load_config(self):
        """Simulasi memuat file eksternal yang memicu KonfigurasiError jika berkas tidak ditemukan."""
        try:
            # Skenario simulasi: Jika nama file bukan 'config.json', anggap FileNotFoundError terjadi [cite: 738]
            if self.file_config != "config.json":
                raise FileNotFoundError(f"Berkas fisik '{self.file_config}' tidak ditemukan pada direktori aktif.")

            # Jika berkas ditemukan, isi konfigurasi aktif aplikasi
            self._config = {"tema": "dark", "bahasa": "id", "volume": "80"}
            print(f"[System] Berhasil mengonfigurasi aplikasi dari '{self.file_config}'")

        except FileNotFoundError as e_file:
            # Membungkus FileNotFoundError ke dalam custom exception KonfigurasiError [cite: 1224]
            raise KonfigurasiError("Gagal menginisialisasi konfigurasi awal aplikasi.", penyebab_asli=e_file)

    def get_setting(self, key: str, default: str = None) -> str:
        """Mengambil data setting aplikasi; memicu KonfigurasiError jika properti/key salah."""
        try:
            if key not in self._config:
                raise KeyError(f"Properti data '{key}' tidak terdaftar pada skema konfigurasi.")
            return self._config[key]
        except KeyError as e_key:
            # Membungkus KeyError ke dalam custom exception KonfigurasiError [cite: 1224]
            raise KonfigurasiError(f"Gagal memuat preferensi pengguna pada properti: '{key}'", penyebab_asli=e_key)


# Pengujian Praktikum 06
if __name__ == "__main__":
    print("\n=== RUNNING: PRAKTIKUM 06 ===")

    # Skenario A: Inisialisasi Berkas Salah (Memicu FileNotFoundError -> KonfigurasiError)
    print("--- Skenario Berkas Mengalami Kegagalan ---")
    try:
        app_gagal = PengelolaSetting("app_preferences.ini")
    except KonfigurasiError as error_aplikasi:
        print(f"Pesan Eksepsi Ditangkap:\n -> {error_aplikasi}")

    # Skenario B: Berkas Benar, Memanggil Properti yang Salah (Memicu KeyError -> KonfigurasiError)
    print("\n--- Skenario Properti Mengalami Kegagalan ---")
    try:
        app_sukses = PengelolaSetting("config.json")
        # Pemanggilan sukses
        print(f" -> Pilihan Bahasa: {app_sukses.get_setting('bahasa')}")
        # Pemanggilan memicu kegagalan
        print(f" -> Ukuran Resolusi: {app_sukses.get_setting('resolusi_layar')}")
    except KonfigurasiError as error_aplikasi:
        print(f"Pesan Eksepsi Ditangkap:\n -> {error_aplikasi}")
```

```python
class InventarisBarang:
    def __init__(self):
        # a) Inisialisasi atribut protected (_items) sebagai dictionary kosong
        self._items = {}  # Format data -> {kode_barang: jumlah_stok}

    def tambah_barang(self, kode, jumlah):
        """
        b) Menambahkan jumlah stok untuk kode barang tertentu.
        - Memastikan jumlah adalah integer positif (ValueError).
        - Memicu KeyError jika kode barang sudah ada (Anggap tambah hanya untuk barang baru).
        """
        print(f"\n[Metode tambah_barang] Mencoba mendaftarkan '{kode}' dengan stok: {jumlah}")
        try:
            # Validasi tipe data dan nilai integer positif
            if not isinstance(jumlah, int) or jumlah <= 0:
                raise ValueError(f"Jumlah harus berupa integer positif! Input tidak valid: {jumlah}")

            # Validasi duplikasi key barang baru
            if kode in self._items:
                raise KeyError(f"Error: Kode barang '{kode}' sudah ada.")

            # Jika valid dan belum ada, masukkan ke dictionary
            self._items[kode] = jumlah
            print(f" -> Sukses: Barang '{kode}' berhasil ditambahkan dengan stok {jumlah}.")

        except ValueError as e_val:
            print(f" -> ValueError ditangkap di dalam metode: {e_val}")
        # Catatan: KeyError sengaja tidak ditangkap di sini agar di-raise ke kode utama (sesuai poin e)

    def ambil_barang(self, kode, jumlah):
        """
        c) Mengurangi jumlah stok untuk kode barang tertentu.
        - Memastikan jumlah adalah integer positif (ValueError).
        - Menangani jika kode barang tidak ditemukan (KeyError).
        - Memicu ValueError jika jumlah yang diminta > stok yang tersedia.
        """
        print(f"\n[Metode ambil_barang] Mencoba mengambil '{kode}' sejumlah: {jumlah}")
        try:
            # 1. Validasi nilai integer positif
            if not isinstance(jumlah, int) or jumlah <= 0:
                raise ValueError(f"Jumlah penyerahan harus berupa integer positif! Input: {jumlah}")

            # 2. Validasi keberadaan kode barang
            if kode not in self._items:
                raise KeyError(f"Error: Kode barang '{kode}' tidak ditemukan.")

            # 3. Validasi kecukupan stok barang
            stok_sekarang = self._items[kode]
            if jumlah > stok_sekarang:
                raise ValueError(f"Error: Stok barang '{kode}' tidak mencukupi! (tersedia: {stok_sekarang}, diminta: {jumlah}).")

            # Jika semua lolos, kurangi stok fisik internal
            self._items[kode] -= jumlah
            print(f" -> Sukses: Berhasil mengambil {jumlah} pcs '{kode}'. Sisa stok saat ini: {self._items[kode]} pcs.")

        except KeyError as e_key:
            print(f" -> KeyError ditangkap di dalam metode: {e_key}")
        except ValueError as e_val:
            # Menangkap ValueError internal (baik dari input negatif maupun stok tidak cukup)
            print(f" -> ValueError ditangkap di dalam metode: {e_val}")
            # Memicu kembali raise ValueError khusus stok kurang ke kode utama jika diperlukan
            if "tidak mencukupi" in str(e_val):
                raise

    def cek_stok(self, kode):
        """
        d) Memeriksa dan mencetak stok menggunakan struktur lengkap:
        try...except KeyError...else...finally
        """
        print(f"\n[Metode cek_stok] Memeriksa status barang '{kode}':")
        try:
            # try: Coba akses dictionary internal
            stok = self._items[kode]
        except KeyError:
            # except KeyError: Cetak pesan jika barang tidak ada
            print(f" -> Hasil: Barang dengan kode '{kode}' TIDAK ditemukan dalam sistem.")
        else:
            # else: Dijalankan jika try berhasil (tanpa error)
            print(f" -> Hasil: Kode '{kode}' TERSEDIA dengan total stok: {stok} pcs.")
        finally:
            # finally: Selalu dieksekusi di akhir proses pengecekan
            print(f" -> [System] Pengecekan stok untuk kode '{kode}' telah selesai dilakukan.")


# e) Bagian Kode Utama Pengujian Sistem (__name__ == "__main__")
if __name__ == "__main__":
    print("====================================================")
    print("      PENGUJIAN SISTEM MANAJEMEN INVENTARIS BARANG   ")
    print("====================================================")

    # Inisialisasi objek utama
    gudang = InventarisBarang()

    # --- PENGUJIAN OPRASI TAMBAH BARANG ---
    print("\n>>> 1. OPERASI: TAMBAH BARANG")
    gudang.tambah_barang("BRG-01", 50)  # Sukses
    gudang.tambah_barang("BRG-02", -10) # Memicu ValueError (Ditangkap di internal metode)

    # Kasus Tambah kode sama (Memicu KeyError yang dilempar/raise ke kode utama)
    try:
        gudang.tambah_barang("BRG-01", 20)
    except KeyError as error_luar:
        print(f" -> [Main Code Catch] Exception KeyError tertangkap di kode utama: {error_luar}")


    # --- PENGUJIAN OPERASI AMBIL BARANG ---
    print("\n>>> 2. OPERASI: AMBIL BARANG")

    # Kasus A: Kode tidak ada (Ditangkap KeyError di dalam metode)
    gudang.ambil_barang("BRG-UNKNOWN", 5)

    # Kasus B: Jumlah tidak valid / negatif (Ditangkap ValueError di dalam metode)
    gudang.ambil_barang("BRG-01", -5)

    # Kasus C: Stok tidak cukup (Memicu ValueError yang di-raise ke kode utama)
    try:
        gudang.ambil_barang("BRG-01", 100)
    except ValueError as error_stok:
        print(f" -> [Main Code Catch] Exception ValueError tertangkap di kode utama: {error_stok}")

    # Kasus D: Pengambilan sukses normal
    gudang.ambil_barang("BRG-01", 15)


    # --- PENGUJIAN OPERASI CEK STOK (TRY...EXCEPT...ELSE...FINALLY) ---
    print("\n>>> 3. OPERASI: CEK STOK")

    # Cek kode yang terdaftar (Memicu Blok Else & Finally)
    gudang.cek_stok("BRG-01")

    # Cek kode yang tidak terdaftar (Memicu Blok Except & Finally)
    gudang.cek_stok("BRG-99")
```
